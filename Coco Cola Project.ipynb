```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Coca-Cola_stock_history (1).csv')
```

```python
data
```

```python
data.index
```

```python
data.info()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Open', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Open")
plt.xlabel('Open')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Close', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Close")
plt.xlabel('Close')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'High', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of High")
plt.xlabel('High')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Low', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Low")
plt.xlabel('Low')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Volume', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Volume")
plt.xlabel('Volume')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Dividends', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Dividends")
plt.xlabel('Dividends')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Stock Splits', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Stock Splits")
plt.xlabel('Stock Splits')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Low', color = 'blue',
             edgecolor = 'linen', alpha = 0.5, bins = 5)

plt.title("Histogram of Low")
plt.xlabel('Low')
plt.ylabel('Count')
plt.vlines(data['Low'].mean(), ymin = 0, ymax = 4000, colors='blue', label='Mean')
plt.vlines(data['Low'].median(), ymin = 0, ymax = 4000, colors='red', label='Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Volume', color = 'blue',
             edgecolor = 'linen', alpha = 0.5, bins = 5)

plt.title("Histogram of Volume")
plt.xlabel('Volume')
plt.ylabel('Count')
plt.vlines(data['Volume'].mean(), ymin = 0, ymax = 4000, colors='blue', label='Mean')
plt.vlines(data['Volume'].median(), ymin = 0, ymax = 4000, colors='red', label='Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.distplot(data['Close'], color = 'blue')

plt.title("Distplot of Close")
plt.xlabel('Close')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.distplot(data['Dividends'], color = 'blue')

plt.title("Distplot of Dividends")
plt.xlabel('Dividends')
plt.ylabel('Density')
plt.show()
```
